```python
df_list = pd.read_csv('list.csv',header= None)
names = list(df_list.values.flatten())
```

```python
names
```

```python
df=pd.DataFrame()
for i in range(len(names)):
    a = (names[i])
    try:
        df = get_history(symbol=a, start=date(2005,1,1), end=date(2018,1,1))
        df = df.append(df)
        print(i) #can remove this line, just prints the index for the list of names
    except:
        df = df
```
